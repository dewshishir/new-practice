```python
img = np.linspace(0, 255, 40000, dtype=np.uint8).reshape(200, 200)

# apply same threshold (230) with different max_val
_, th0   = cv2.threshold(img, 230, 0,   cv2.THRESH_BINARY_INV)
_, th4   = cv2.threshold(img, 230, 4,   cv2.THRESH_BINARY_INV)
_, th140 = cv2.threshold(img, 230, 140, cv2.THRESH_BINARY_INV)
_, th255 = cv2.threshold(img, 230, 255, cv2.THRESH_BINARY_INV)

# show all in one figure
fig, axs = plt.subplots(1, 4, figsize=(14, 4))

axs[0].imshow(th0, cmap='gray', vmin=0, vmax=255)
axs[0].set_title('max_val = 0 (all black)')

axs[1].imshow(th4, cmap='gray', vmin=0, vmax=255)
axs[1].set_title('max_val = 4 (dim gray)')

axs[2].imshow(th140, cmap='gray', vmin=0, vmax=255)
axs[2].set_title('max_val = 140 (medium gray)')

axs[3].imshow(th255, cmap='gray', vmin=0, vmax=255)
axs[3].set_title('max_val = 255 (bright white)')

for ax in axs:
    ax.axis('off')

plt.tight_layout()
plt.show()

# print unique pixel values for verification
print("Unique pixel values:")
print("max_val = 0   :", np.unique(th0))
print("max_val = 4   :", np.unique(th4))
print("max_val = 140 :", np.unique(th140))
print("max_val = 255 :", np.unique(th255))
```

Pixels ≤ 60 → unchanged (0–60)

Pixels > 60 → clipped to 60

Auto-scale: Matplotlib rescales the brightest pixel (here 60) → white

Result: whole image appears bright/white (because max = 60 scaled to 255)

```python
ret, thresholded = cv2.threshold(img,60,255,cv2.THRESH_TRUNC)
#lebur protita dot ekta pixel
# All pixels ≤220 → unchanged

# All pixels >220 → become 220
plt.imshow(thresholded,cmap='gray')
plt.show(ret)
```

Pixels ≤ 20 → unchanged (0–20)

Pixels > 20 → clipped to 20

Fixed scale: vmin=0, vmax=255

Pixel values now are 0–20 out of 0–255 → very dark

Result: image looks black

```python
ret, thresholded = cv2.threshold(img,20,255,cv2.THRESH_TRUNC)
#lebur protita dot ekta pixel
# All pixels ≤220 → unchanged

# All pixels >220 → become 220
plt.imshow(thresholded, cmap='gray', vmin=0, vmax=255)
plt.show(ret)
```
