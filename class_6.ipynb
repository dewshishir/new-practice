```python
 adaptive_mean = cv.adaptiveThreshold(img,255,cv.ADAPTIVE_THRESH_MEAN_C,cv.THRESH_BINARY,5,10) #eikhaen c hccche fine control er jnno

 plt.imshow(adaptive_mean,cmap='gray')
```

```python
adaptive_15 = cv.adaptiveThreshold(img, 255,
                                   cv.ADAPTIVE_THRESH_MEAN_C,
                                   cv.THRESH_BINARY,
                                   15, 2)

adaptive_3 = cv.adaptiveThreshold(img, 255,
                                  cv.ADAPTIVE_THRESH_MEAN_C,
                                  cv.THRESH_BINARY,
                                  3, 2)

# Show both side by side
plt.figure(figsize=(10,5))

plt.subplot(1, 2, 1)
plt.imshow(adaptive_15, cmap='gray')
plt.title('blockSize = 15')

plt.subplot(1, 2, 2)
plt.imshow(adaptive_3, cmap='gray')
plt.title('blockSize = 3')

plt.show()
```

```python
ret1, otsul = cv.threshold(img, 160, 255, cv.THRESH_BINARY + cv.THRESH_OTSU)
ret2, otsul2 = cv.threshold(img, 60, 255, cv.THRESH_BINARY + cv.THRESH_OTSU)
plt.figure(figsize=(10,5))
plt.subplot(1, 2, 1)
plt.imshow(otsul, cmap='gray')
plt.title('threshold = 160')

plt.subplot(1, 2, 2)
plt.imshow(otsul2, cmap='gray')
plt.title('threshold = 60')

plt.show()

# eikhane output er thersold value same jodio amra same dei nai eikhaner thersold value kto ta ber korte hbe?
```
