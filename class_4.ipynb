# Rotation

```python
rotated_image = cv2.rotate(image, cv2.ROTATE_90_CLOCKWISE) #90 * clock wise
#display and rotate image
plt.figure(figsize=(10,5))
plt.subplot(1,2,1) #dui image plot korar jnno use hoi (ekta row,dui ta coloum,image 1 number e thakbe)
plt.imshow(image)
plt.title('original image')
plt.subplot(1,2,2) #oitar ektu vinno
plt.imshow(rotated_image)
plt.title('rotated image')
plt.show()
```

```python
image.shape
```

```python
# Get the image dimensions
(h, w) = image.shape[:2]

# Calculate the center of the image
center = (w // 2, h // 2)

# Create a rotation matrix
M1 = cv2.getRotationMatrix2D(center, 30, 1.0)
M2 = cv2.getRotationMatrix2D(center, 45, 1.0)
M3 = cv2.getRotationMatrix2D(center, -90, 0.6)


# Rotate the image
rotated_image_30 = cv2.warpAffine(image, M1, (w, h))
rotated_image_45 = cv2.warpAffine(image, M2, (w, h))
rotated_image_60 = cv2.warpAffine(image, M3, (w, h))


plt.figure(figsize=(15,5))

plt.subplot (131)
plt.imshow(rotated_image_30)
plt.title( '30 degree rotated')

plt.subplot (132)
plt.imshow(rotated_image_45)
plt.title( '45 degree rotated')

plt.subplot (133)
plt.imshow(rotated_image_60)
plt.title( '90 degree rotated')

plt.tight_layout()
plt.show()
```
